# GABA-B Synapse


```python
# Add code here
# to start a new build to run multiple full simulations
device.reinit()
device.activate()
# need to reset the defaultclock after activate
defaultclock.dt = 20 * us

network = Network()
chip = DynapSE(network)
seed(4321)
# Create a poisson spike generator with 100 Hz
spike_generator = PoissonGroup(1, 10*Hz)
# Allocate single neuron from Core 1
DPI_neuron  = chip.get_neurons(1, 'Core_1')
DPI_neuron.Isoma_const = 80 * pA #(5nA)#2000 * pA

# Define an NMDA synapse to connect the spike generator to the neuron 
DPI_synapse = chip.add_connection(spike_generator, DPI_neuron, synapse_type='GABA_A')#GABA_A
# DPI_synapseA = chip.add_connection(spike_generator, DPI_neuron, synapse_type='GABA_A')#GABA_A
# DPI_synapse2 = chip.add_connection(spike_generator, DPI_neuron, synapse_type='AMPA')#GABA_A
# In Brian2 creating a Synapses instance does not create synapses, it only specifies their dynamics.
# so remember to call connect function to really build the connection
chip.connect(DPI_synapse, True)
# chip.connect(DPI_synapse2, True)

# here we set an initial weight of 200.
DPI_synapse.weight = 300
# DPI_synapse2.weight = 300

# ampa_conf = {"I_tau_syn_ampa": 10 * pA}

gaba_conf = {"Igaba_a_tau": 10 * pA}#, "Igaba_b_g": 50 * pA}
# ampa_conf = {"Iampa_tau": 50 * pA}#, "Iampa_g": 50 * pA}

chip.set_bias(gaba_conf, 'Core_1')
# chip.set_bias({**gaba_conf,**ampa_conf}, 'Core_1')

# monitors
mon_synapse_state = StateMonitor(DPI_synapse, 'Igaba_a', record=[0]) # it's not I_syn_exc2
# mon_synapse_state2 = StateMonitor(DPI_synapse2, 'Iampa', record=[0]) # it's not I_syn_exc2
mon_neuron_state  = StateMonitor(DPI_neuron,  'Isoma_mem', record=True)
mon_neuron_input  = SpikeMonitor(spike_generator, name='mon_neuron_input')
mon_neuron_output = SpikeMonitor(DPI_neuron, name='mon_neuron_output')

# add all the instances into the network
network.add([spike_generator, DPI_neuron, DPI_synapse,mon_synapse_state,#DPI_synapse2 mon_synapse_state2, 
            mon_neuron_state, mon_neuron_input, mon_neuron_output])

#DPI_synapse,mon_synapse_state
# Simulation
duration = .1 * second
network.run(duration)
print('Output firing rate = ', mon_neuron_output.count/ (duration))
```

```python
# plotting
plt.figure(figsize=(16,5),dpi=150)
plt.subplots_adjust(hspace=.5)

times = int(duration/ms)

# Input Spikes 
plt.subplot(311)
spike_placeholder = np.zeros(times)
spike_placeholder[np.array(mon_neuron_input.t/ms, dtype=int)]=1
plt.plot(spike_placeholder,'k|')
matplotlib.pyplot.yticks(range(1, 3))
plt.ylim(0.5, 1.5)
plt.title('Input Spikes')
plt.ylabel('Neuron ID')
#plt.xlabel('Time (ms)')

plt.subplot(312)
Isoma_mem = mon_neuron_state.Isoma_mem[0]
plt.plot(mon_neuron_state.t/ms, Isoma_mem/pA)
# plt.hlines(DPI_synapse.Inmda.get_item(0)/pA,0,100,'k',linestyles='dotted')
plt.ylim(-5., 300)

plt.title('Neuron Membrane Current')
plt.ylabel('Current (pA)')
#plt.xlabel('Time (ms)')

plt.subplot(313)
spike_placeholder = np.zeros(times)
spike_placeholder[np.array(mon_neuron_output.t/ms, dtype=int)]=1
plt.plot(spike_placeholder,'k|')
matplotlib.pyplot.yticks(range(1, 3))
plt.ylim(0.5, 1.5)
plt.title('Neuron Output Spikes')
plt.ylabel('Neuron ID')
plt.xlabel('Time (ms)')
plt.show()
```

```python
plt.figure(figsize=(16,3),dpi=150)

Igaba_a = mon_synapse_state.Igaba_a[0]
plt.plot(mon_synapse_state.t/ms, Igaba_a/nA)

plt.title('Synaptic Current')
plt.ylabel('Current (nA)')
plt.xlabel('time (ms)')
```
