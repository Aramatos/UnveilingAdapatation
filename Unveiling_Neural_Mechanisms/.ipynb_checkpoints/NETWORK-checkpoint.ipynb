```python
from brian2 import *
from DynapSE import DynapSE
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

# Display plots inside Jupyter cell
%matplotlib inline 
# Set the dots-per-inch (resolution) of the images
mpl.rcParams['figure.dpi'] = 90
from equations.dynapse_eq import *
from parameters.dynapse_param import *

# C++ code generation for faster spiking network simulation
set_device('cpp_standalone')
# Ignore Brian2 base warnings
BrianLogger.suppress_name('base')
# The clock of Brian2 simulation for numerically solve ODEs
defaultclock.dt = 20 * us
```

```python
input_type   = 'regular' # 'regular', 'poisson' or 'cosine' (Default: poisson)
pulse_start  = 0         # second - Start time of input (Default: 0)
pulse_stop   = 2    # second - Stop time of input (Default: 5)
inp_duration = 4        # second - Simulation duration (Default: 5)
rate         = 50 # Hz or rad/sec - Spiking rate (Default: 80 Hz for regular, 100 Hz for poission, 2 rad/sec for cosine)


spikes = np.zeros(inp_duration*1000)
dt = int(1000/rate)
spikes[pulse_start*1000:pulse_stop*1000:dt] = 1.0

plt.plot(spikes,'k|')
plt.xlabel("Time (ms)")
plt.ylabel("Neuron ID")
plt.ylim(0.5, 1.5)
plt.show()

spike_timing = np.where(spikes==1)[0] * ms # Timing of spikes
neuron_indices = np.zeros(len(spike_timing)) # ID of spike sources

```

```python
# Add code here
# to start a new build to run multiple full simulations
device.reinit()
device.activate()
# need to reset the defaultclock after activate
defaultclock.dt = 20 * us


Cue = SpikeGeneratorGroup(1, indices=neuron_indices, times=spike_timing, name='InpSpikeGenerator')    

network = Network()
chip = DynapSE(network)
seed(4321)
# Create a poisson spike generator with 100 Hz
Stimulus = PoissonGroup(1, 100*Hz)

# Allocate single neuron from Core 1
Neuron1 = chip.get_neurons(1, 'Core_1')
Neuron2 = chip.get_neurons(1, 'Core_2')
Neuron3 = chip.get_neurons(1, 'Core_3')
Neuron1.Isoma_const = 0 * pA #(5nA)#2000 * pA
Neuron2.Isoma_const = 0 * pA #(5nA)#2000 * pA
Neuron3.Isoma_const = 0 * pA #(5nA)#2000 * pA

# Define an NMDA synapse to connect the spike generator to the neuron 
DPI_synapse1 = chip.add_connection(Cue, Neuron1, synapse_type='AMPA') #core 1
DPI_synapse2 = chip.add_connection(Neuron1, Neuron2, synapse_type='GABA_A') #core 2
DPI_synapse3 = chip.add_connection(Neuron2, Neuron3, synapse_type='GABA_A') #core 3
DPI_synapse4 = chip.add_connection(Stimulus, Neuron2, synapse_type='AMPA')#core 2
DPI_synapse5 = chip.add_connection(Stimulus, Neuron3, synapse_type='AMPA')#core 3
# In Brian2 creating a Synapses instance does not create synapses, it only specifies their dynamics.
# so remember to call connect function to really build the connection
chip.connect(DPI_synapse1, True)
chip.connect(DPI_synapse2, True)
chip.connect(DPI_synapse3, True)
chip.connect(DPI_synapse4, True)
chip.connect(DPI_synapse5, True)

# here we set an initial weight of 200.
DPI_synapse1.weight = 300
DPI_synapse2.weight = 300
DPI_synapse3.weight = 2000
DPI_synapse4.weight = 100
DPI_synapse5.weight = 50

DPI_synapse5.delay = 10*ms

gaba_conf = {"Igaba_a_tau": 5 * pA}#, "Igaba_b_g": 50 * pA} DOSENT MATTER
ampa_conf = {"Iampa_tau": 10 * pA}#, "Iampa_g": 50 * pA} CUE TO NEURON1
chip.set_bias(gaba_conf, 'Core_1')
chip.set_bias(ampa_conf, 'Core_1')

gaba_conf = {"Igaba_a_tau": 30* pA}#, "Igaba_b_g": 50 * pA} Neuron 1 to Neuron 2
ampa_conf = {"Iampa_tau": 20* pA}#, "Iampa_g": 50 * pA} Stim to Neuron 2
chip.set_bias(gaba_conf, 'Core_2')
chip.set_bias(ampa_conf, 'Core_2')

gaba_conf = {"Igaba_a_tau": 120 * pA}#, "Igaba_b_g": 50 * pA} Neuron 2 to Neuron 3
ampa_conf = {"Iampa_tau": 15 * pA}#, "Iampa_g": 50 * pA} Stim to Neuron 3
chip.set_bias(gaba_conf, 'Core_3')
chip.set_bias(ampa_conf, 'Core_3')


# monitors
mon_synapse1_state = StateMonitor(DPI_synapse1, 'Iampa', record=[0])
mon_synapse2_state = StateMonitor(DPI_synapse2, 'Igaba_a', record=[0])
mon_synapse3_state = StateMonitor(DPI_synapse3, 'Igaba_a', record=[0])
mon_synapse4_state = StateMonitor(DPI_synapse4, 'Iampa', record=[0])
mon_synapse5_state = StateMonitor(DPI_synapse5, 'Iampa', record=[0])


mon_neuron_state1  = StateMonitor(Neuron1,  'Isoma_mem', record=True)
mon_neuron_state2  = StateMonitor(Neuron2,  'Isoma_mem', record=True)
mon_neuron_state3  = StateMonitor(Neuron3,  'Isoma_mem', record=True)

mon_Stimulus  = SpikeMonitor(Stimulus, name='mon_Stimulus')
mon_Cue = SpikeMonitor(Cue, name='mono_Cue')

mon_neuron_output = SpikeMonitor(Neuron3, name='mon_neuron_output')

# add all the instances into the network
network.add([Cue,Stimulus])
network.add([Neuron1,Neuron2,Neuron3,DPI_synapse1,DPI_synapse2,DPI_synapse3,DPI_synapse4,DPI_synapse5])
network.add([mon_synapse1_state,mon_synapse2_state,mon_synapse3_state,mon_synapse4_state,mon_synapse5_state])
network.add([mon_neuron_state1,mon_neuron_state2,mon_neuron_state3,mon_Stimulus,mon_Cue,mon_neuron_output])

network.run(inp_duration*second)
```

```python
# Plotting

tstart =0# ms 
tend   = 4000 # ms 
s = int(1*ms / defaultclock.dt)

#input subplots
# plotting
plt.figure(figsize=(16,5),dpi=150)
plt.subplots_adjust(hspace=.5)

times = int(inp_duration/ms)

# Input Spikes 
plt.subplot(311)
spike_placeholder = np.zeros(times)
spike_placeholder[np.array(mon_Cue.t/ms, dtype=int)]=1
plt.plot(spike_placeholder,'k|')
matplotlib.pyplot.yticks(range(0, 3))
plt.ylim(0.5, 3)
plt.title('Cue')
plt.ylabel('Neuron ID')

plt.subplot(312)
spike_placeholder = np.zeros(times)
spike_placeholder[np.array(mon_Stimulus.t/ms, dtype=int)]=1
plt.plot(spike_placeholder,'k|')
matplotlib.pyplot.yticks(range(0, 3))
plt.ylim(0.5, 3)
plt.title('Stimulus')
plt.ylabel('Neuron ID')

plt.subplot(313)
spike_placeholder = np.zeros(times)
spike_placeholder[np.array(mon_neuron_output.t/ms, dtype=int)]=1
plt.plot(spike_placeholder,'k|')
matplotlib.pyplot.yticks(range(0, 3))
plt.ylim(0.5, 3)
plt.title('Stimulus')
plt.ylabel('Neuron ID')


fig2, (ax1,ax2,ax3) = plt.subplots(1, 3,figsize=(20,3))

ax1.plot(mon_neuron_state1.t[tstart*s:tend*s]/ms, mon_neuron_state1.Isoma_mem[0][tstart*s:tend*s], linewidth=1.5)
ax1.legend(['Imem_'+str(1)])
ax1.set_title('Inhibitor1')
ax1.set_ylabel('Current (A)')
ax1.set_xlabel('Time (ms)')
ax1.grid(True)

ax2.plot(mon_neuron_state2.t[tstart*s:tend*s]/ms, mon_neuron_state2.Isoma_mem[0][tstart*s:tend*s], linewidth=1.5)
ax2.legend(['Imem_'+str(2)])
ax2.set_title('Inhibitor2')
ax2.set_xlabel('Time (ms)')
ax2.grid(True)

ax3.plot(mon_neuron_state3.t[tstart*s:tend*s]/ms, mon_neuron_state3.Isoma_mem[0][tstart*s:tend*s], linewidth=1.5)
ax3.legend(['Imem_'+str(3)])
ax3.set_title('OUTPUT')
ax3.set_ylabel('Current (A)')
ax3.set_xlabel('Time (ms)')
ax3.grid(True)

fig2, (ax1,ax2,ax3,ax4,ax5) = plt.subplots(1, 5,figsize=(20,3))

ax1.plot(mon_synapse1_state.t[tstart*s:tend*s]/ms, mon_synapse1_state.Iampa[0][tstart*s:tend*s], linewidth=1.5)
ax1.set_title('Cue to N1')
ax1.set_ylabel('Current (A)')
ax1.set_xlabel('Time (ms)')
ax1.grid(True)

ax2.plot(mon_synapse2_state.t[tstart*s:tend*s]/ms, mon_synapse2_state.Igaba_a[0][tstart*s:tend*s], linewidth=1.5)
ax2.set_title('N1 to N2')
ax2.set_xlabel('Time (ms)')
ax2.grid(True)

ax3.plot(mon_synapse3_state.t[tstart*s:tend*s]/ms, mon_synapse3_state.Igaba_a[0][tstart*s:tend*s], linewidth=1.5)
ax3.set_title('N2 to N3')
ax3.set_xlabel('Time (ms)')
ax3.grid(True)

ax4.plot(mon_synapse4_state.t[tstart*s:tend*s]/ms, mon_synapse4_state.Iampa[0][tstart*s:tend*s], linewidth=1.5)
ax4.set_title('Stim to N2')
ax4.set_xlabel('Time (ms)')
ax4.grid(True)

ax5.plot(mon_synapse5_state.t[tstart*s:tend*s]/ms, mon_synapse5_state.Iampa[0][tstart*s:tend*s], linewidth=1.5)
ax5.set_title('Stim to N3')
ax5.set_xlabel('Time (ms)')
ax5.grid(True)
```
